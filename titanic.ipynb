# Q6.) categories of people and ages.

```python
df.groupby('Age').Sex.value_counts()
```

```python
def people(age):
    if age <15:
        return 'child'
    elif age >=15 and age < 25:
        return 'Tenager'
    elif age >=25 and  age <50:
        return 'middle age'
    elif age>=50:
        return 'older'
```

```python
df['people'] =  df['Age'].apply(people)
```

```python
df['people'].value_counts()
```

```python
sns.countplot(df['people'])
```
